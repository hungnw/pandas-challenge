```python
import pandas as pd

path = "Resources/purchase_data.csv"
df = pd.read_csv(path)
df.head()
```

# PLAYER COUNT

```python
total_players = len(df["SN"].value_counts())
p = pd.DataFrame({"Total Players":[total_players]})
p
```

# PURCHASE ANALYSIS (TOTAL)

```python
unique_items = len(df["Item ID"].value_counts())
average_price = df["Price"].mean()
number_of_purchase = df["Purchase ID"].count()
total_revenue = df["Price"].sum()

pat = pd.DataFrame({"Number of Unique Items":[unique_items],"Average Purchase Price":[average_price],
                   "Total Number of Purchase":[number_of_purchase],"Total Revenue":[total_revenue]})
pat['Average Purchase Price'] = pd.Series(["${0:.2f}".format(val) for val in 
                                                pat['Average Purchase Price']], index = pat.index)
pat['Total Revenue'] = pd.Series(["${0:.2f}".format(val) for val in 
                                                pat['Total Revenue']], index = pat.index)

pat
```

# GENDER DEMOGRAPHICS

```python
g = df[["Gender","SN"]].drop_duplicates()
g = g.groupby("Gender").count()
g = g.rename(columns={"SN":"Total Count"})
g["Percentage of Players"] = g["Total Count"]/total_players
g['Percentage of Players'] = pd.Series(["{0:.2f}%".format(val * 100) for val in g['Percentage of Players'
                                                                                   ]], index = g.index)
g.sort_values("Total Count",ascending=False)
```

# PURCHASE ANALYSIS (GENDER)

```python
pa1 = df[["Gender","Purchase ID"]].groupby("Gender").count()
pa1 = pa1.rename(columns={"Purchase ID":"Purchase Count"})

pa2 = df[["Gender","Price"]].groupby("Gender").mean()
pa2 = pa2.rename(columns={"Price":"Average Purchase Price"})

pa3 = df[["Gender","Price"]].groupby("Gender").sum()
pa3 = pa3.rename(columns={"Price":"Total Purchase Value"})

pa3["Average Purchase per Person"] = pa3["Total Purchase Value"]/g["Total Count"]

pa_m1 = pd.merge(pa1,pa2,on="Gender")

pa_m2 = pd.merge(pa_m1,pa3,on="Gender")

pa_m2
```

# AGE DEMOGRAPHICS

```python
bins = [0,9.5,14.5,19.5,24.5,29.5,34.5,39.5,99]
age_groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39",">40"]
df["Age Groups"]=pd.cut(df["Age"], bins, labels=age_groups) 

a = df[["SN","Age Groups"]].drop_duplicates()

a1 = a.groupby("Age Groups").count()
a1 = a1.rename(columns={"SN":"Total Count"})

total_players = a["SN"].count()
a1["Percentage of Players"]=a1["Total Count"]/total_players
a1['Percentage of Players'] = pd.Series(["{0:.2f}%".format(val * 100) for val in a1['Percentage of Players'
                                                                                   ]], index = a1.index)
a1
```

# PURCHASING ANALYSIS (AGE)

```python
pc1 = df[["Age Groups","Purchase ID"]].groupby("Age Groups").count()
pc1 = pc1.rename(columns={"Purchase ID":"Purchase Count"})


pc2 = df[["Age Groups","Price"]].groupby("Age Groups").mean()
pc2 = pc2.rename(columns={"Price":"Average Purchase Price"})

pc3 = df[["Age Groups","Price"]].groupby("Age Groups").sum()
pc3 = pc3.rename(columns={"Price":"Total Purchase Value"})

pc3["Average Total Purchase per Person"]=pc3["Total Purchase Value"]/a1["Total Count"]

pc_m = pd.merge(pc1,pc2,on="Age Groups")

pc_m2 = pd.merge(pc_m,pc3,on="Age Groups")

pc_m2
```

# TOP SPENDERS

```python
s1 = df[["SN","Purchase ID"]].groupby("SN").count()

s2 = df[["SN","Price"]].groupby("SN").mean()

s3 = df[["SN","Price"]].groupby("SN").sum()

sm1 = pd.merge(s1,s2,on="SN")

sm2 = pd.merge(sm1,s3,on="SN")

sm3 = sm2.rename(columns={"Purchase ID":"Purchase Count","Price_x":"Average Purchase Price",
                         "Price_y":"Total Purchase Value"})

sm3.sort_values("Total Purchase Value",ascending=False).head()
```

# MOST POPULAR ITEMS

```python
p1 = df[["Item ID","Item Name","Purchase ID"]].groupby(["Item ID","Item Name"]).count()
p1 = p1.rename(columns={"Purchase ID":"Purchase Count"})

p2 = df[["Item ID","Item Name","Price"]].drop_duplicates()

p3 = df[["Item ID","Item Name","Price"]].groupby(["Item ID","Item Name"]).sum()
p3 = p3.rename(columns={"Price":"Total Purchase Value"})

pm1 = pd.merge(p1,p2,on=["Item ID","Item Name"])

pm2 = pd.merge(pm1,p3,on=["Item ID","Item Name"])
pm2.sort_values("Purchase Count",ascending=False).head()
```
